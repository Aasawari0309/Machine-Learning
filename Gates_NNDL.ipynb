```python
#AND Gate
#single layer perceptron
import numpy as np
#define unit step function
def unitStep(v):
  if v>=0:
    return 1
  else:
    return 0
```

```python
#design perceptron model
def perceptronModel(x,w,b):
  v=np.dot(x,w)+b
  y=unitStep(v)
  return y
```

```python
#w1=1,w2=1,b=-1.5
def And_function(x):
  w=np.array([1,1])
  b=-1.5
  return perceptronModel(x,w,b)
```

```python
#Testing
result=np.array([1,0])
print("AND({},{})={} ".format(1,0,And_function(result)))
```
