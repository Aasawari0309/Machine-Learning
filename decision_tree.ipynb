```python
#import libraries and dataset
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.model_selection import train_test_split

data=sns.load_dataset('iris')
data.head()
```

```python
#separating the training set and testing set
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30)
x_train2,x_validation,y_train2,y_validation=train_test_split(x_train,y_train,test_size=0.30)
#import decision tree classifier via sklearn
from sklearn.tree import DecisionTreeClassifier

#Applying Decision Tree classifier
dtree = DecisionTreeClassifier(criterion='entropy',max_depth=3)
dtree.fit(x_train2,y_train2)
predictions=dtree.predict(x_validation)
predict_test=dtree.predict(x_test)
#checking accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_validation,predictions)
accuracy_score(y_test,predict_test)
```

```python
dtree2 = DecisionTreeClassifier(criterion='gini',max_depth=3)
dtree2.fit(x_train2,y_train2)
predictions2=dtree2.predict(x_validation)
predict_test2=dtree2.predict(x_test)
#checking accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_validation,predictions2)
accuracy_score(y_test,predict_test2)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig=plt.figure(figsize=(16,12))
a=plot_tree(dtree,feature_names=data.columns,fontsize=12,filled=True,class_names=['Setosa','Versicolor','Virginica'])
```
