```python
#import the libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#load the iris data
data=sns.load_dataset('iris')
data.head()
```

```python
#prepare the training data for buiding the model
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

```python
#Plot the relation of each feature with each species
plt.xlabel('Features')
plt.ylabel('Species')

pltX=data.loc[:,'sepal_length']
pltY=data.loc[:,'species']
plt.scatter(pltX,pltY,color='b',label='sepal_length')

pltX=data.loc[:,'sepal_width']
pltY=data.loc[:,'species']
plt.scatter(pltX,pltY,color='g',label='sepal_width')

pltX=data.loc[:,'petal_length']
pltY=data.loc[:,'species']
plt.scatter(pltX,pltY,color='r',label='petal_length')

pltX=data.loc[:,'petal_width']
pltY=data.loc[:,'species']
plt.scatter(pltX,pltY,color='indigo',label='petal_width')
plt.legend(loc=4,prop={'size':8})

plt.show()
```

```python
#split the dataset
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
#Create and train logistic regression
model=LogisticRegression()
model.fit(x_train,y_train)
LogisticRegression()
```

```python
#test the model
predictions = model.predict(x_test)
print(predictions)
```

```python
#Check precision,recall,f1-score
print(classification_report(y_test,predictions))
```

```python
print(accuracy_score(y_test,predictions))
```
